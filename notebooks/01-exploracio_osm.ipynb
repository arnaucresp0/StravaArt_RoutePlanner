```python
# %%
# Cel·la 1: Configuració inicial actualitzada
import osmnx as ox
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from IPython.display import display, HTML

# Configuració per versions noves d'OSMnx
try:
    # Per versions >= 1.0
    ox.settings.use_cache = True
    ox.settings.log_console = True
    print(f"Versió OSMnx: {ox.__version__} - Utilitzant settings")
except AttributeError:
    # Per versions antigues (< 1.0)
    ox.config(use_cache=True, log_console=True)
    print(f"Versió OSMnx: {ox.__version__} - Utilitzant config")

# Configuració de visualització
%matplotlib inline
plt.rcParams['figure.figsize'] = [12, 10]
plt.rcParams['figure.dpi'] = 100

print("✅ Entorn configurat correctament!")
```

```python
# %%
# Cel·la 2: Prova amb zona petita de Manresa
print("Carregant xarxa de carrers...")

# Zona petita per provar (Plaça Catalunya, Manresa)
G = ox.graph_from_address("Plaça Catalunya, Manresa", 
                         dist=500, network_type='walk')

print(f"Nodes: {len(G.nodes())}, Vores: {len(G.edges())}")
```

```python
# %%
# Cel·la 3: Visualitza la xarxa
fig, ax = plt.subplots(figsize=(10, 10))
ox.plot_graph(G, ax=ax, node_size=0, edge_linewidth=0.5, show=False)
ax.set_title("Xarxa de carrers - Plaça Catalunya, Manresa")
plt.show()
```

```python
# %%
# Cel·la 4: Explora les dades
# Primer node
first_node = list(G.nodes())[0]
print(f"Node d'exemple: {first_node}")
print(f"Posició (lat, lon): {G.nodes[first_node]}")

# Algunes vores
print(f"\nVores d'exemple:")
for i, edge in enumerate(list(G.edges())[:3]):
    print(f"  {edge}")
```
